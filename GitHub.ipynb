### 1- Introduction.
GitHub est une plateforme destiné à toutes personnes pour héberger leurs code, leurs poésie, leurs Texte,... il est développé en Ruby on Rails et Erlang par **Chris Wanstrath, PJ Hyett et Tom Preston-Werner** en 2008.

Le nom GitHub est composé de :  Git et Hub.

  - **Git**: C'est le coeur de GitHUB, développé par Linus Torvalds.
  Il s'agit d'un **système de contrôle de version open-source**,il gère les modifications d'un projet sans 
  écraser n'importe quelle partie ;
  
  - **Hub**: le mot HUB à plusieurs significations, mais dans ce cas, il s'agit d'un réseau social  bâti autour de Git, mais aussi à une plate-forme de correspondance.
 
 en bref : GitHub est un espace d'hébergement des projets pour le controle de version et le Travail en collaboration.
 
 Avec Git, on utilise des lignes de commandes ça sera peut être  une chose non préférable pour les personnes qui aiment les interfaces visuelles en créant des comptes sur **GitHUB.com**.
 
 En avril 2016, GitHub a annoncé avoir dépassé les 14 millions d'utilisateurs et plus de 35 millions de dépôts de projets, aujourd'hui il comporte 40 millions utilisateurs.

 Le 4 juin 2018, **Satya Nadella**, le Directeur Général (PDG) de Microsoft, annonce que son entreprise **Microsoft** est dans un processus d'acquisition de GitHub pour 7,5 milliards de dollars.

### 3- Principaux constituants de GitHub (Worflow) : 

 - 1) **Référentiel (Repository)**:
    Espace d'hébergement de notre projet.

 - 2) **Les branches**:
    * La première chose à faire est de créer une nouvelle branche.
    * La branche est une réplique exacte de l branche master.
    * c'est une méthode sécurisée pour apporter des modifications sans affecter la production.
 
 - 3) **Les Validations(commit)**:
   * un commit est une modification/ changement .
   * le fait de faire un commit implique l'ouverture d'une demande sur GitHub.
 
 - 4) **Les requêtes de tirage(Pull Request)**:
    * comparer la branche crée avec la branche "master".
    * montrer les différences aux autres développeurs.
    
 - 5) **Collaborate**:
   * les développeurs voient les changements, et on peut collaborer avec eux.
   * les développeurs peuvent accepter/ critiquer et donner des suggestions. 
 
 - 6) **Merge**
   * fusionner les travaux en un seul travail.
   * remettre en "master" la branche crée.

### Travail avec Git :  

 **Etape 1:**
 - Créer un compte GitHub
 - Vérifier est ce que Git est installé :
      - **git --version**
  s'il affiche un message d'erreur, il faut [intaller git](https://git-scm.com/)
  
   
  
 **Etape 2:** Définir l'identité de l'utilisateur :
  - git config --global user.name "name_user_GitHub"
  - git config --global user.email "compte_email"
  
 **Etape 3:** Création d'un projet en local
 * Créer un répertoire : **mkdir nom_dossier**
 * Entrer dans le répertoire : **cd nom_dossier/**
 * Initialiser git dans le projet : **git init**
 * Voir à tout moment le status d'un fichier : **git status**
 * Ecrire le fichier à programmer en le sauvegarder dans mon_dossier.
 * Selectionner des fichiers : **git add FileName**
 
 **Etape 4:** Faire des commits en local
 * **git commit -m "first commit"**
 A chaque modification on va sélectionner des fichiers :
 * **git add FileName**
 * voir la date et le nom d'utilisateur des commits :**git log**
 
 **Etape 5:** Les branches
 * Créer une branche : **git branch name_branch**
 * Lister les branches : **git branch**
 * Accéder à la branche : **git checkout name_branch**
 * Faire des commits (Etape 4)
 * Basculer vers la branche master :
   - Sélectionner la branche principale : **git checkout master**
   - Fusionner les commits : **git merge name_branch**
   - Suppression de la branche: **git branch -d name_branch**
   - Lister : **git branch**
