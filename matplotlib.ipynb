# Matplotlib

Matplotlib est une bibliothèque python open source destinée pour la représentation graphique des données.

Elle est écrit initialement par John D.Hunter (1968-2012), aujourd'hui elle est développée par d'autres personnes.

### Installation de matplotlib : 

Il existe différentes façons pour installer matplotlib, parmis elles on trouve :

- pip install matplotlib.

### Importer matplotlib :

- import matplotlib.pyplot as plt 

ou avec ipython 

- In[1] : % matplotlib

- In[2] : import matplotlib.pyplot as plt 

### Représentation graphique :

Il existe différents graphes pour représenter nos données , on va les résumer sous forme d'un tableau : 

|Type de tracé | Fonction |
| :---: | :---: |
|Ligne|plt.plot()|
|Barres vérticales|plt.bar()|
|Barres horizontales|plt.barh()|
|histogrammes|plt.hist()|
|boxplot|plt.box()|
|pacelles de surface|plt.area()|
|nuages de points|plt.scatter()|
|parcelles à tarte|plt.pie()|
|parcelles bin hexagonales|plt.hexbin()|

### Remarques importantes : 
 En apprentissage machine deux méthodes existent : **Pyplot** et **OOP** 


#### 1- Pyplot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [10,12,20,56,30,42]
plt.plot(x,y)
```

```python
x = [1,2,3,4,5,6]
y = [10,12,20,56,30,42]
plt.bar(x,y)
```

```python
x = [1,2,3,4,5,6]
y = [10,12,20,56,30,42]
plt.barh(x,y)
```

```python
x = [1,2,3,4,5,6]
y = [10,12,20,56,30,42]
plt.hist(x)
```

```python
x = [1,2,3,4,5,6]
y = [10,12,20,56,30,42]
plt.scatter(x,y)
```

```python
x = [1,2,3,4,5,6]
y = [10,12,20,56,30,42]
plt.hexbin(x,y)
```
